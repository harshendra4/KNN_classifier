## Providing labels to our text data

```python
df4= df3.copy()
df4.ST_Slope.replace({'Up':1,'Flat':2,'Down':3}, inplace=True)
df4.ExerciseAngina.replace({'N':0,'Y':1}, inplace=True)
df4.RestingECG.replace({'Normal':1, 'ST':2, 'LVH':3}, inplace=True)
df4.ChestPainType.replace({'ATA':1, 'NAP':2, 'ASY':3, 'TA':4}, inplace=True)
```

```python
df4
```

```python
Dummies=pd.get_dummies(df4.Sex)
```

```python
Dummies
```

```python
Dummies=Dummies.drop(['F'],axis='columns')
```

```python
merged=pd.concat([df4, Dummies], axis='columns')
```

```python
merged
```

```python
merged.drop(['Sex'],axis='columns')
```

```python
X= merged.drop(['HeartDisease','Sex'], axis='columns')
X
```

```python
y=merged.HeartDisease
```

```python
y
```

## Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_scaled= scaler.fit_transform(X)
X_scaled
```

## Splitting the data into test and train sample

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.svm import SVC
svm= SVC()
svm.fit(X_train, y_train)
```

```python
svm.score(X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lg= LogisticRegression()
lg.fit(X_train, y_train)
```

```python
lg.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=80)
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_test, y_test)
```

```python
X
```

## Using PCA and giving argument that come up with 98% useful data from our input data

```python
from sklearn.decomposition import PCA
pca= PCA(0.98)
X_pca= pca.fit_transform(X)
```

```python
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
df=pd.DataFrame(X_train_pca)
```

```python
df
```

```python
lg1= LogisticRegression()
lg1.fit(X_train_pca, y_train)
```

```python
lg1.score(X_test_pca, y_test)
```

```python
rfc1=RandomForestClassifier(n_estimators=80)
rfc1.fit(X_train_pca, y_train)
```

```python
rfc1.score(X_test_pca, y_test)
```

### PCA didn't improved the score, it means here we lose the accuracy on reducing the dimensions
